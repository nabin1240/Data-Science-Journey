```python
# Topic (Multiple Line Chart, Colors, Markets, Grid, Legend, Save Chart as Image Advaned Pie Chart)
#lesson 1 line chart
import matplotlib.pyplot as plt
days = [1,2,3,4,5]
sales = [100,200,150,300,250]
profit = [20,40, 30, 50, 45]
plt.plot(days, sales, label="Sales")
plt.plot(days, profit, label="Profit")
plt.title("Sales va Profit")
plt.xlabel("Days")
plt.ylabel("Amount")
plt.legend()
plt.savefig("Chart.png")
plt.show()
```

```python
plt.plot(days, sales, color="green")  #color lesson 2 in color we can add more colors like (blue,green, red, orange)
```

```python
plt.plot(days, sales, marker="o")
```

```python
# Lesson 4 Grid 
plt.plot(days, sales, marker="o")
plt.grid()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]

sale = [20,30,80,50,30,70,20]

profit = [2000,300,500,1000,700,3000,400]

plt.figure(figsize=(10,5))

plt.plot(day, sale,
label="sale",
marker="o",
color="green")

plt.plot(day, profit,
label="profit",
marker="o",
color="blue")

plt.grid()

plt.legend()

plt.title("Sell vs Profit")

plt.xlabel("Day")

plt.ylabel("Amount")

plt.show()
```

```python
import matplotlib.pyplot as plt
app = ["TikTok", "Facebook", "Youtube"]
users = [40,30,20]
plt.pie(users, labels=app,autopct="%1.1f%%")
plt.title("App Usage")
plt.show()
```

```python
#lesson 8 __ Professional Stock Chart
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("data.xlsx")
top10 = df.head(10)
plt.figure(figsize=(12,5))
plt.plot(top10["Symbol"], top10["LTP"], marker="o", color="green", label="Stock Price")
plt.title("Top 10 Stock Prices")
plt.xlabel("Symbol")
plt.ylabel("LTP")
plt.xticks(rotation=90)
plt.grid()
plt.legend()
plt.show()
```

```python
#Homework
import pandas as pd
import matplotlib.pyplot as plt
Day = [1,2,3,4,5,6,7]
Sales = [200,300,100,500,150,200,100]
Profit = [1000,3000,700,1000,8000,400,100]

plt.figure(figsize=(12,5))
plt.plot(Day, Sales, label="Sales", marker="o", color="blue")

plt.plot(Day, Profit, label="Profit", marker="o", color="green")
plt.title("Top Sales Vs Profit")
plt.xlabel("Day")
plt.ylabel("Amount")
plt.grid()
plt.legend()
plt.savefig("chart.png")
plt.show()
```
