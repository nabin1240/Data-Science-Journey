```python
#Lesson 1 - How many rows are available in dataset
import pandas as pd
df = pd.read_excel("data.xlsx")
df = df.dropna(axis=1 , how="all")
print("Total Rows:", df.shape[0])
```

```python
#lesson 2 - Mean 
#Average LTP
print("Average LTP:", df["LTP"].mean())
```

```python
#Lesson 3 - sum
# Total Volume
print("Total Volume:", df["Volume"].sum())
```

```python
#Multiple Statistics
print(df["LTP"].describe())
```

```python
# Lesson  5 - Create New Column 
# Convert Volume into New formate
df["Volume_k"] = df["Volume"] / 1000
print(df[["Symbol", "Volume", "Volume_k"]].head())
```

```python
#Lesson 6 - Top 3 Hightest LTP
top3 = df.sort_values("LTP", ascending=False)
print(top3[["Symbol", "LTP"]].head(3))
```

```python
# Lesson 7 - Top 3 Lowest LTP
top3_lowest = df.sort_values("LTP", ascending=True)
print(top3_lowest[["Symbol", "LTP"]].head(3))
```
