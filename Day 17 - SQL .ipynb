```python
#Step 2 - Table Create
import sqlite3
conn = sqlite3.connect("school.db")
cursor = conn.cursor()
cursor.execute(""" CREATE TABLE students (
id INTGER,
name TEXT,
age INTEGER
)
""")
conn.commit()
print("Table Created")
```

```python
# Step 3 - insert Data
cursor.execute("""
INSERT INTO students
VALUES(1,'Ram',20)
"""
              )
conn.commit()
```

```python
# Multiple Data
cursor.execute(
"""
INSERT INTO students
VALUES (2,'Hari',22)
"""
)
cursor.execute(
""" 
INSERT INTO students 
VALUES (3, 'Sits',19)
""")
conn.commit()
```

```python
#Step - Select 
cursor.execute(
"""
SELECT * FROM students
"""
)
print(cursor.fetchall())
```

```python
#Step 5 _ Specific Column
cursor.execute("""
SELECT name FROM students
""")
print(cursor.fetchall())
```

```python
#Step ^ _ WHERE 
cursor.execute(
    """ SELECT * FROM students 
    WHERE age > 20
    """
)
print(cursor.fetchall())
```

```python
# Step 7 - ORDER BY  
cursor.execute(
    """
    SELECT * FROM students ORDER BY age DESC 
    """
)
print(cursor.fetchall())
```

```python
# Step 8 - LIMIT
cursor.execute(
    """ SELECT * FROM students
    LIMIT 2
    """

)
print(cursor.fetchall())
```

```python
import sqlite3

conn = sqlite3.connect("Stock2.db")

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS Stock (
    Id INTEGER,
    symbol TEXT,
    ltp INTEGER
)
""")

cursor.execute("INSERT INTO Stock VALUES (1,'ADBL',317)")
cursor.execute("INSERT INTO Stock VALUES (2,'AHPC',287)")
cursor.execute("INSERT INTO Stock VALUES (3,'ANLB',6038)")
cursor.execute("INSERT INTO Stock VALUES (4,'API',359)")

conn.commit()

cursor.execute("SELECT * FROM Stock")

print(cursor.fetchall())

conn.close()
```
