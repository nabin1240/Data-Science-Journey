```python
import pandas as pd
df = pd.read_excel("DAY 21.xlsx")
df = df[["ID", "NAME", "SALARY"]].dropna()
print(df)
print("Total Values:\n", df["SALARY"].sum, "Average Values:\n", df["SALARY"].mean(), "Max Values : \n", df["SALARY"].max(), "Lowest Values: \n", df["SALARY"].min(), "Top 1 worker ", df.loc[df["SALARY"].idxmax()])
```

```python
import matplotlib.pyplot as plt
plt.bar(
    df["NAME"], 
    df["SALARY"]
)
plt.title("Workers Data")
plt.xlabel("Workes")
plt.ylabel("Salary")
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

```python
#Import In sql
import sqlite3
conn = sqlite3.connect("workers.db")
df.to_sql("workers",conn, if_exists="replace", index=False)
print("saved")
sql_workers_data = pd.read_sql_query("SELECT * FROM workers",conn)
print(sql_workers_data)
```
