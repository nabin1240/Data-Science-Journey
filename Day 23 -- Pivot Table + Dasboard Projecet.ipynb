```python
import pandas as pd
data = { 
    "Stock" : ["Nabil", "SBI", "NMB", "SHIVM", "UPPER", "NIFRA"],
    "Sector" : ["BANK", "BANK", "BANK", "CEMENT", "HYDRO", "FINANCE"],
    "Price" : [650,420,580,670,203,350]
}
df = pd.DataFrame(data)
pivot = df.groupby("Sector")["Price"].sum()
print(pivot)
                   
```

```python
# Homework
data = {
    "Name" : ["Nabin", "Ram", "Hari", "Sita"],
    "Department" : ["IT", "HR", "IT", "Finance"],
    "Salary": [30000,40000,60000,200000]
}
df = pd.DataFrame(data)
pivot_tb = df.groupby("Department")["Salary"].sum()
print(pivot_tb)
```

```python
#Highest Salary 
Highest = df.loc[df["Salary"].idxmax()]
print(Highest)
```

```python
#Homework
import pandas as pd
worker = {
    "Name" : ["Nabin" , "Lahan", "Krishna", "Hari", "Govinda", "Ram"],
    "Group" : ["IT", "IT", "Agriculture", "IT", "Agriculture", "Plumber"],
    "Salary" :[250000,246000,50000,45500,67000,50000]
}
df = pd.DataFrame(worker)
pivot_gp = df.groupby("Group")["Salary"].sum()
print(pivot_gp)
#Average
print("Average Salary: \n",df.groupby("Group")["Salary"].mean())

#Maximum Salary
print("Maximum Salary :\n",df.groupby("Group")["Salary"].max())
#Minimum Salary
print("Mini Salary: \n", df.groupby("Group") ["Salary"].min())
#Count Salary
print("Total Count :\n", df.groupby("Group") ["Salary"].count())
    
```
