```python
#Data Frame
import sqlite3
import pandas as pd
conn = sqlite3.connect("company.db")
df = pd.read_sql_query("SELECT* FROM Employees",conn )
print(df)
```

```python
# Data frame Analysis in pandas
print("Max: ",df["Salary"].max()) 
```

```python
# total salary
print("Total Salary:",df["Salary"].sum())
```

```python
#Average Salary
print(" Average Salary:", df["Salary"].mean())
```

```python
# Top Salary Employee
print("Top Salary:",df.loc[df["Salary"].idxmax()])
```

```python
# Salary > 40000
high_salary = df[df["Salary"]> 40000 ]
print(high_salary)
```

```python
# sort value High to Low
print(df.sort_values("Salary", ascending=False))
```

```python
# Low to High Values
print(df.sort_values("Salary", ascending=True))
```

```python
# Save to Excel file (employees.xlsx)
df.to_excel("employees.xlsx",index=False)
print("Saved")
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("Stock_data.db")
cursor = conn.cursor()

# Create Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS stock_data (
    Id INTEGER,
    Name TEXT,
    Price INTEGER
)
""")

# Insert Data
cursor.execute("INSERT INTO stock_data VALUES (1, 'Nabil', 3000)")
cursor.execute("INSERT INTO stock_data VALUES (2, 'SBI', 500)")
cursor.execute("INSERT INTO stock_data VALUES (3, 'NMB', 450)")
cursor.execute("INSERT INTO stock_data VALUES (4, 'NIMB', 340)")
cursor.execute("INSERT INTO stock_data VALUES (5, 'NIFRA', 350)")

conn.commit()

# Load to DataFrame
df = pd.read_sql_query(
    "SELECT * FROM stock_data",
    conn
)

print(df)

# Highest Price
print("Max Price:", df["Price"].max())

# Average Price
print("Average Price:", df["Price"].mean())

# Price > 200
price = df[df["Price"] > 200]
print(price)

# High to Low
sort_values = df.sort_values(
    "Price",
    ascending=False
)
print(sort_values)

# Low to High
lowest_price = df.sort_values(
    "Price",
    ascending=True
)
print(lowest_price)

# Top 1 Price
top_price = df.loc[
    df["Price"].idxmax()
]
print(top_price)

conn.close()
```
