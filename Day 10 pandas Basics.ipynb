```python
#Create DataFrame
import pandas as pd   
data = {
    "Name" : ["Nabin", "Ram", "Hari"],
    "age" : [22, 21, 23]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Access Column
print(df["Name"])
```

```python
# Basic Analysis
print(df["age"].mean())
print(df["age"].max())
print(df["age"].min())
```

```python
# Check Data Type
print(df.dtypes)
```
