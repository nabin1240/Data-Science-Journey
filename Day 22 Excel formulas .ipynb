```python
import pandas as pd
data = {
    "Name" : ["Nabin", "Ram", "Hari"],
    "Salary" : [30000, 50000, 45000]
}
df = pd. DataFrame(data)
print(df)
df["Status"] = df["Salary"].apply(
    lambda x : "High" if x > 40000 else "Low")
print(df)
```
