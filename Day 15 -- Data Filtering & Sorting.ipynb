```python
# Top Gainer Finding Here Lesson 1

import pandas as pd
df = pd.read_excel("data.xlsx")
df = df.dropna(axis=1, how="all")
top_gainer= df.sort_values("% Change", ascending=False)
print(top_gainer[["Symbol", "% Change"]].head())
```

```python
# Top Loser Lesson 2
top_loser = df.sort_values("% Change", ascending=True)
print(top_loser[["Symbol", "% Change", ]].head())
```

```python
# Highest Volume Stock Lesson 3
import matplotlib.pyplot as plt
top_volume = df.sort_values("Volume", ascending=False)
print(top_volume[["Symbol", "Volume"]].head())
plt.plot(top_volume["Symbol"], top_volume["Volume"])
plt.title("Highest Volume Stock")
plt.xlabel("Symbol")
plt.ylabel("Volume")
plt.xticks(rotation=90)
plt.show()
```

```python
# Data Short Lesson 4
# Finding Grater than 300 + LTP Stock 
high_ltp = df[df["LTP"]>300]
print(high_ltp[["Symbol", "LTP"]])
```

```python
#Multiple Conditon 
result = df[
    (df["LTP"] > 300) & 
    (df["Volume"] > 10000) ]
print(result[["Symbol", "LTP", "Volume"]])
```

```python
#Lesson 6 -- Single Stock Search
#Example : ADBL
stock = df[df["Symbol"] == "ADBL"]
print(stock)
```

```python
# Lesson 7 - Summary Statistics
print(df[["Symbol","LTP"]].max())
print(df[["Symbol","LTP"]].min())
print(df["LTP"].mean())
```
