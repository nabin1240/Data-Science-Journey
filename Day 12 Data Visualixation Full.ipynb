```python
#Lesson 1 Line chart
import matplotlib.pyplot as plt
days = [1,2,3,4,5]
sales = [100,200,150,300,250]
plt.plot(days,sales)
plt.title("Sales Trend")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
# bar Chart
students = ["Ram", "Hari", "Nabin"]
marks = [8, 90, 75]
plt.figure(figsize=(10,5))
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.xticks(rotation=90)
plt.show()
           
```

```python
# Pie Chart
apps = ["TikTok", "Facebook", "YouTube"]
users = [40, 30, 30]
plt.pie(users, labels=apps)
plt.title("App Usage")
plt.show()
```

```python
# Histogram Chart
ages = [20,21,22,23,24,25,20,21,22]
plt.hist(ages)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
        
```
